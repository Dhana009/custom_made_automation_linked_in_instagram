```python
import cv2
import numpy as np
import calendar
import pyperclip as p
import os
import datetime
from PIL import Image, ImageDraw, ImageFont

class monthtracker :
    
    def month(self):
        # Get user input for year, month, and date
        year = int(input("Enter the year: "))
        month = int(input("Enter the month (1-12): "))
        date = int(input("Enter the date: "))

        # Check if a custom template image path is provided
        template_image_path = input("Enter the path of the template image (leave blank for basic template): ")

        # Load the template image or create a basic template
        if template_image_path:
            template_image = cv2.imread(template_image_path)
        else:
            template_image = np.ones((1200, 1000, 3), dtype=np.uint8) * 255

        # Create a copy of the template image
        image = template_image.copy()

        # Set the font properties
        font = cv2.FONT_HERSHEY_SIMPLEX
        font_scale = 1.5
        font_thickness = 2
        text_color = (0, 0, 0)

        # Calculate the size and position of each date cell
        cell_size = (image.shape[1] - 2) // 7
        cell_x = 1
        cell_y = 200

        # Draw the month and year
        month_name = calendar.month_name[month]
        year_text = str(year)
        text_size_month, _ = cv2.getTextSize(month_name, font, font_scale, font_thickness)
        text_size_year, _ = cv2.getTextSize(year_text, font, font_scale, font_thickness)
        text_x = (image.shape[1] - text_size_month[0] - text_size_year[0]) // 2
        text_y = 100
        cv2.putText(image, month_name, (text_x, text_y), font, font_scale, text_color, font_thickness)
        cv2.putText(image, year_text, (text_x + text_size_month[0] + 10, text_y), font, font_scale, text_color, font_thickness)

        # Calculate the starting position of the dates in the calendar grid
        first_day_weekday, num_days = calendar.monthrange(year, month)
        start_position = (first_day_weekday - 1) % 7

        # Draw the dates
        for day in range(1, num_days + 1):
            # Get the position of the date in the calendar grid
            row = (day + start_position - 1) // 7
            col = (day + start_position - 1) % 7

            # Calculate the position of the date cell
            cell_x = 1 + col * cell_size
            cell_y = 200 + row * cell_size

            if day == date:
                # Get the rating for the selected date
                rating = int(input("Rate the working progress for date {} (0-5): ".format(day)))

                # Calculate the rating color based on the rating value
                rating_color = (0, 255, 0)  # Default to green
                if rating == 0:
                    rating_color = (0, 0, 0)  # Black color for 0 rating
                elif rating > 0 and rating <= 5:
                    rating_color = (0, 255 * rating // 5, 0)  # Gradually increase green intensity

                # Draw the rating color for the date cell
                cv2.rectangle(image, (cell_x, cell_y), (cell_x + cell_size, cell_y + cell_size), rating_color, -1)

            # Draw the small edge between squares
            #cv2.rectangle(image, (cell_x, cell_y), (cell_x + cell_size, cell_y + cell_size), (0, 0, 0), 0)

            # Draw the date text in the center of the square
            text_size, _ = cv2.getTextSize(str(day), font, font_scale, font_thickness)
            text_x = cell_x + (cell_size - text_size[0]) // 2
            text_y = cell_y + (cell_size + text_size[1]) // 2
            cv2.putText(image, str(day), (text_x, text_y), font, font_scale, text_color, font_thickness)

        # Display the calendar image
        cv2.imshow("Calendar", image)
        cv2.waitKey(0)
        cv2.destroyAllWindows()

        # Save the calendar image with a unique filename
        filename = "{}_{}_{}.png".format(calendar.month_abbr[month], year, date)
        os.chdir(r'C:\Users\dhana\Pictures\time tracker for month')
        print(p.copy(os.getcwd() +"\\" +filename))
        print(os.getcwd() +"\\" +filename)
        cv2.imwrite(filename, image)
        print("Calendar image saved as:", filename)
        
        

class sessionimagescreator:
    
    def sessionimages(self):
        ask = int(input('enter the session no 1 - 3  for output :'))
        xx_coordiante = int(input("enter any rating number from 282 to 740 -- rating 0 for 282 -- / 1 - 344 / 2 - 450 / 3 -550 / 4 - 650 / 5 -740"))
        topic = input('enter the topic of this session : ')
        duration = input('enter the duration of this session : ')
        
        for i in range(1,3):
            image_path = r"C:\Users\dhana\Pictures\Add a subheading" + "\\" + f"{i}.png"  # Replace with the path to your image
            image = Image.open(image_path)

            # Create a copy of the image to preserve the original
            image_with_elements = image.copy()

            # Set the font properties
            font_path = r"C:\Users\dhana\Pictures\Add a subheading\Poppins-ExtraBold.ttf"  # Replace with the path to your font file

            font_color = (255, 0, 0)  # Red color (RGB)
            background_color = (255, 255, 255)  # Blue color (RGB)

            # Create a drawing object
            draw = ImageDraw.Draw(image_with_elements)

            # Get the current date and day
            current_date = datetime.date.today().strftime("%B %d, %Y")
            current_day = datetime.datetime.now().strftime("%A")
            text = f"{current_day} {current_date} With DJ"

            if i==1:
                font_size = 33
                # Prompt for input
                text = text
                x = 290
                y = 790

                # Calculate text size
                font = ImageFont.truetype(font_path, font_size)
                text_width, text_height = draw.textsize(text, font=font)

                # Calculate background rectangle coordinates
                background_x = x
                background_y = y
                background_width = text_width
                background_height = text_height

                # Draw the background rectangle
                draw.rectangle((background_x, background_y, background_x + background_width, background_y + background_height), fill=background_color)

                # Add the custom text to the image
                draw.text((x, y), text, font=font, fill=font_color)
                image_with_elements.show()
                # Save the image with added text
                output_path = r"C:\Users\dhana\Pictures\session_emotions"+"\\"+f"{i}.png"  # Replace with the desired path and filename
                image_with_elements.save(output_path)
        
            elif i == 2:
                font_color = (0, 0, 0)
                font_size_date = 25
                x_date = 11
                y_date = 14

                font = ImageFont.truetype(font_path, font_size_date)
                # Calculate text size
                text_width, text_height = draw.textsize(text, font=font)
                draw.text((x_date,y_date), text, font=font, fill=font_color)

                text = topic
                font_size_date = 32
                x_date = 610
                y_date = 62

                # Calculate text size
                text_width, text_height = draw.textsize(text, font=font)
                draw.text((x_date, y_date), text, font=font, fill=font_color)

                font_size = 65
                # Prompt for input
                image = Image.open(r"C:\Users\dhana\Pictures\Add a subheading\arrow.png")
                text = "<▼>"
                x = xx_coordiante
                y = 260

                font = ImageFont.truetype(font_path, font_size)
                # Add the custom text to the image
                draw.text((x, y), text, font=font, fill=font_color)

                # Display the image with added text
                image_with_elements.show()
                # Save the image with added text
                output_path = r"C:\Users\dhana\Pictures\session_emotions"+"\\"+f"{i}.png"  # Replace with the desired path and filename
                image_with_elements.save(output_path)

                for ii in range(1,5):
                    images = Image.open(r"C:\Users\dhana\Pictures\Add a subheading\session"+"\\"+f"{ii}.png")
                    print(images)
                    image_with_elements = images.copy()
                      # Replace with the path to your font file

                    if ii==1:
                        draw = ImageDraw.Draw(image_with_elements)

                        font_path = r"C:\Users\dhana\Pictures\Add a subheading\Poppins-ExtraBold.ttf"
                        font_color = (255, 255, 255)
                        font_size_date = 35
                        text = f"{current_date}"
                        durations = f"{duration}hrs"
                        x_date,y_date = 11,14
                        timedurationx,timedurationy = 11,65
                        sessionx,sessiony = 845,530

                        lstlist = [[x_date,y_date],[timedurationx,timedurationy],[sessionx,sessiony]]
                        textformat = [text,durations,ask]

                        for i in range(0,3):
                            if i!=2:
                                font = ImageFont.truetype(font_path, font_size_date)
                                # Calculate text size
                                text_width, text_height = draw.textsize(text, font=font)
                                draw.text((lstlist[i][0],lstlist[i][1]), str(textformat[i]), font=font, fill=font_color)
                            if i==2:
                                font_size_date = 80
                                font = ImageFont.truetype(font_path, font_size_date)
                                # Calculate text size
                                text_width, text_height = draw.textsize(text, font=font)
                                draw.text((lstlist[i][0],lstlist[i][1]), str(textformat[i]), font=font, fill=font_color)

                        # Save the image with added text
                        output_path = r"C:\Users\dhana\Pictures\session_emotions"+"\\"+"3.png"  # Replace with the desired path and filename
                        image_with_elements.save(output_path)
                    if ii==2:
                        draw = ImageDraw.Draw(image_with_elements)

                        font_path = r"C:\Users\dhana\Pictures\Add a subheading\Poppins-ExtraBold.ttf"
                        font_color = (0, 0, 0)
                        font_size_date = 75
                        session = f" Topic : {topic}"
                        takeaway = input('enter the key take aways 1 for this topic :')
                        x_date,y_date = 130,121
                        timedurationx,timedurationy = 110,400


                        lstlist = [[x_date,y_date],[timedurationx,timedurationy]]
                        textformat = [session,takeaway]

                        for i in range(0, 2):
                            if i != 2:

                                font_size_date = 63
                                font = ImageFont.truetype(font_path, font_size_date)
                                # Calculate text size
                                text_width, text_height = draw.textsize(textformat[i], font=font)

                                max_width = 800  # Adjust the maximum width as needed

                                if text_width > max_width:
                                    words = textformat[i].split()
                                    lines = []
                                    current_line = ""
                                    for word in words:
                                        test_line = current_line + " " + word if current_line != "" else word
                                        test_width, _ = draw.textsize(test_line, font=font)
                                        if test_width <= max_width:
                                            current_line = test_line
                                        else:
                                            lines.append(current_line)
                                            current_line = word
                                    lines.append(current_line)

                                    line_y = lstlist[i][1]
                                    for line in lines:
                                        font_color = (255, 255 ,255)
                                        draw.text((lstlist[i][0], line_y), line, font=font, fill=font_color)
                                        line_y += text_height + 5  # Adjust the line spacing as needed
                                else:
                                    font_color = (0, 0, 0)
                                    draw.text((lstlist[i][0], lstlist[i][1]), str(textformat[i]), font=font, fill=font_color)
                        # Save the image with added text
                        output_path = r"C:\Users\dhana\Pictures\session_emotions"+"\\"+"4.png"  # Replace with the desired path and filename
                        image_with_elements.save(output_path)
                    if ii==3:

                        draw = ImageDraw.Draw(image_with_elements)

                        font_path = r"C:\Users\dhana\Pictures\Add a subheading\Poppins-ExtraBold.ttf"

                        font_size_date = 75
                        session = f" Topic  :  {topic}"
                        takeaway = input('enter the key take aways 2 for this topic :')
                        x_date,y_date = 130,121
                        timedurationx,timedurationy = 110,450


                        lstlist = [[x_date,y_date],[timedurationx,timedurationy]]
                        textformat = [session,takeaway]

                        for i in range(0, 2):
                            if i != 2:
                                font_size_date = 60
                                font = ImageFont.truetype(font_path, font_size_date)
                                # Calculate text size
                                text_width, text_height = draw.textsize(textformat[i], font=font)

                                max_width = 800  # Adjust the maximum width as needed

                                if text_width > max_width:
                                    words = textformat[i].split()
                                    lines = []
                                    current_line = ""
                                    for word in words:
                                        font_color = (0, 0 ,0)
                                        test_line = current_line + " " + word if current_line != "" else word
                                        test_width, _ = draw.textsize(test_line, font=font)
                                        if test_width <= max_width:
                                            current_line = test_line
                                        else:
                                            lines.append(current_line)
                                            current_line = word
                                    lines.append(current_line)

                                    line_y = lstlist[i][1]
                                    for line in lines:
                                        font_color = (0, 0, 0)
                                        draw.text((lstlist[i][0], line_y), line, font=font, fill=font_color)
                                        line_y += text_height + 5  # Adjust the line spacing as needed
                                else:
                                    font_color = (255, 255 ,255)
                                    draw.text((lstlist[i][0], lstlist[i][1]), str(textformat[i]), font=font, fill=font_color)

                        # Save the image with added text
                        output_path = r"C:\Users\dhana\Pictures\session_emotions"+"\\"+"5.png"  # Replace with the desired path and filename
                        image_with_elements.save(output_path)
                    if ii==4:
                        draw = ImageDraw.Draw(image_with_elements)

                        font_path = r"C:\Users\dhana\Pictures\Add a subheading\Poppins-ExtraBold.ttf"
                        font_color = (0, 0, 0)
                        font_size_date = 70
                        session = f" Topic  :  {topic}"
                        takeaway = input('enter the key take aways 2 for this topic :')
                        x_date,y_date = 130,121
                        timedurationx,timedurationy = 110,450


                        lstlist = [[x_date,y_date],[timedurationx,timedurationy]]
                        textformat = [session,takeaway]

                        for i in range(0, 2):
                            if i != 2:
                                font_size_date = 60
                                font = ImageFont.truetype(font_path, font_size_date)
                                # Calculate text size
                                text_width, text_height = draw.textsize(textformat[i], font=font)

                                max_width = 800  # Adjust the maximum width as needed

                                if text_width > max_width:
                                    words = textformat[i].split()
                                    lines = []
                                    current_line = ""
                                    for word in words:

                                        test_line = current_line + " " + word if current_line != "" else word
                                        test_width, _ = draw.textsize(test_line, font=font)
                                        if test_width <= max_width:
                                            current_line = test_line
                                        else:
                                            lines.append(current_line)
                                            current_line = word
                                    lines.append(current_line)

                                    line_y = lstlist[i][1]
                                    for line in lines:
                                        font_color = (255, 255 ,255)
                                        draw.text((lstlist[i][0], line_y), line, font=font, fill=font_color)
                                        line_y += text_height + 5  # Adjust the line spacing as needed
                                else:
                                    font_color = (0, 0 ,0)
                                    draw.text((lstlist[i][0], lstlist[i][1]), str(textformat[i]), font=font, fill=font_color)



                        # Save the image with added text
                        output_path = r"C:\Users\dhana\Pictures\session_emotions"+"\\"+"6.png"  # Replace with the desired path and filename
                        image_with_elements.save(output_path)

                    # Display the image with added text
                    image_with_elements.show()
                
                        
            else :
                if ask == 3:
                    zfont_color = (0, 0, 0)
                    font_size_date = 25
                    x_date = 11
                    y_date = 14

                    font = ImageFont.truetype(font_path, font_size_date)
                    # Calculate text size
                    text_width, text_height = draw.textsize(text, font=font)
                    draw.text((x_date,y_date), text, font=font, fill=font_color)

                    text = topic
                    font_size_date = 32
                    x_date = 610
                    y_date = 62

                    # Calculate text size
                    text_width, text_height = draw.textsize(text, font=font)
                    draw.text((x_date, y_date), text, font=font, fill=font_color)

                    font_size = 65
                    # Prompt for input
                    image = Image.open(r"C:\Users\dhana\Pictures\Add a subheading\arrow.png")
                    text = "<▼>"
                    x = xx_coordiante
                    y = 260

                    font = ImageFont.truetype(font_path, font_size)
                    # Add the custom text to the image
                    draw.text((x, y), text, font=font, fill=font_color)

                    # Display the image with added text
                    image_with_elements.show()
                    # Save the image with added text
                    output_path = r"C:\Users\dhana\Pictures\session_emotions"+"\\"+f"{i}.png"  # Replace with the desired path and filename
                    image_with_elements.save(output_path)

                    for ii in range(1,5):
                        images = Image.open(r"C:\Users\dhana\Pictures\Add a subheading\session"+"\\"+f"{ii}.png")
                        print(images)
                        image_with_elements = images.copy()
                          # Replace with the path to your font file

                        if ii==1:
                            draw = ImageDraw.Draw(image_with_elements)

                            font_path = r"C:\Users\dhana\Pictures\Add a subheading\Poppins-ExtraBold.ttf"
                            font_color = (255, 255, 255)
                            font_size_date = 35
                            text = f"{current_date}"
                            durations = f"{duration}hrs"
                            x_date,y_date = 11,14
                            timedurationx,timedurationy = 11,65
                            sessionx,sessiony = 845,530

                            lstlist = [[x_date,y_date],[timedurationx,timedurationy],[sessionx,sessiony]]
                            textformat = [text,durations,ask]

                            for i in range(0,3):
                                if i!=2:
                                    font = ImageFont.truetype(font_path, font_size_date)
                                    # Calculate text size
                                    text_width, text_height = draw.textsize(text, font=font)
                                    draw.text((lstlist[i][0],lstlist[i][1]), str(textformat[i]), font=font, fill=font_color)
                                if i==2:
                                    font_size_date = 80
                                    font = ImageFont.truetype(font_path, font_size_date)
                                    # Calculate text size
                                    text_width, text_height = draw.textsize(text, font=font)
                                    draw.text((lstlist[i][0],lstlist[i][1]), str(textformat[i]), font=font, fill=font_color)

                            # Save the image with added text
                            output_path = r"C:\Users\dhana\Pictures\session_emotions"+"\\"+"3.png"  # Replace with the desired path and filename
                            image_with_elements.save(output_path)
                        if ii==2:
                            draw = ImageDraw.Draw(image_with_elements)

                            font_path = r"C:\Users\dhana\Pictures\Add a subheading\Poppins-ExtraBold.ttf"
                            font_color = (0, 0, 0)
                            font_size_date = 75
                            session = f" Topic : {topic}"
                            takeaway = input('enter the key take aways 1 for this topic :')
                            x_date,y_date = 130,121
                            timedurationx,timedurationy = 110,400


                            lstlist = [[x_date,y_date],[timedurationx,timedurationy]]
                            textformat = [session,takeaway]

                            for i in range(0, 2):
                                if i != 2:

                                    font_size_date = 63
                                    font = ImageFont.truetype(font_path, font_size_date)
                                    # Calculate text size
                                    text_width, text_height = draw.textsize(textformat[i], font=font)

                                    max_width = 800  # Adjust the maximum width as needed

                                    if text_width > max_width:
                                        words = textformat[i].split()
                                        lines = []
                                        current_line = ""
                                        for word in words:
                                            test_line = current_line + " " + word if current_line != "" else word
                                            test_width, _ = draw.textsize(test_line, font=font)
                                            if test_width <= max_width:
                                                current_line = test_line
                                            else:
                                                lines.append(current_line)
                                                current_line = word
                                        lines.append(current_line)

                                        line_y = lstlist[i][1]
                                        for line in lines:
                                            font_color = (255, 255 ,255)
                                            draw.text((lstlist[i][0], line_y), line, font=font, fill=font_color)
                                            line_y += text_height + 5  # Adjust the line spacing as needed
                                    else:
                                        font_color = (0, 0, 0)
                                        draw.text((lstlist[i][0], lstlist[i][1]), str(textformat[i]), font=font, fill=font_color)
                            # Save the image with added text
                            output_path = r"C:\Users\dhana\Pictures\session_emotions"+"\\"+"4.png"  # Replace with the desired path and filename
                            image_with_elements.save(output_path)
                        if ii==3:

                            draw = ImageDraw.Draw(image_with_elements)

                            font_path = r"C:\Users\dhana\Pictures\Add a subheading\Poppins-ExtraBold.ttf"

                            font_size_date = 75
                            session = f" Topic  :  {topic}"
                            takeaway = input('enter the key take aways 2 for this topic :')
                            x_date,y_date = 130,121
                            timedurationx,timedurationy = 110,450


                            lstlist = [[x_date,y_date],[timedurationx,timedurationy]]
                            textformat = [session,takeaway]

                            for i in range(0, 2):
                                if i != 2:
                                    font_size_date = 60
                                    font = ImageFont.truetype(font_path, font_size_date)
                                    # Calculate text size
                                    text_width, text_height = draw.textsize(textformat[i], font=font)

                                    max_width = 800  # Adjust the maximum width as needed

                                    if text_width > max_width:
                                        words = textformat[i].split()
                                        lines = []
                                        current_line = ""
                                        for word in words:
                                            font_color = (0, 0 ,0)
                                            test_line = current_line + " " + word if current_line != "" else word
                                            test_width, _ = draw.textsize(test_line, font=font)
                                            if test_width <= max_width:
                                                current_line = test_line
                                            else:
                                                lines.append(current_line)
                                                current_line = word
                                        lines.append(current_line)

                                        line_y = lstlist[i][1]
                                        for line in lines:
                                            font_color = (0, 0, 0)
                                            draw.text((lstlist[i][0], line_y), line, font=font, fill=font_color)
                                            line_y += text_height + 5  # Adjust the line spacing as needed
                                    else:
                                        font_color = (255, 255 ,255)
                                        draw.text((lstlist[i][0], lstlist[i][1]), str(textformat[i]), font=font, fill=font_color)

                            # Save the image with added text
                            output_path = r"C:\Users\dhana\Pictures\session_emotions"+"\\"+"5.png"  # Replace with the desired path and filename
                            image_with_elements.save(output_path)
                        if ii==4:
                            draw = ImageDraw.Draw(image_with_elements)

                            font_path = r"C:\Users\dhana\Pictures\Add a subheading\Poppins-ExtraBold.ttf"
                            font_color = (0, 0, 0)
                            font_size_date = 70
                            session = f" Topic  :  {topic}"
                            takeaway = input('enter the key take aways 2 for this topic :')
                            x_date,y_date = 130,121
                            timedurationx,timedurationy = 110,450


                            lstlist = [[x_date,y_date],[timedurationx,timedurationy]]
                            textformat = [session,takeaway]

                            for i in range(0, 2):
                                if i != 2:
                                    font_size_date = 60
                                    font = ImageFont.truetype(font_path, font_size_date)
                                    # Calculate text size
                                    text_width, text_height = draw.textsize(textformat[i], font=font)

                                    max_width = 800  # Adjust the maximum width as needed

                                    if text_width > max_width:
                                        words = textformat[i].split()
                                        lines = []
                                        current_line = ""
                                        for word in words:

                                            test_line = current_line + " " + word if current_line != "" else word
                                            test_width, _ = draw.textsize(test_line, font=font)
                                            if test_width <= max_width:
                                                current_line = test_line
                                            else:
                                                lines.append(current_line)
                                                current_line = word
                                        lines.append(current_line)

                                        line_y = lstlist[i][1]
                                        for line in lines:
                                            font_color = (255, 255 ,255)
                                            draw.text((lstlist[i][0], line_y), line, font=font, fill=font_color)
                                            line_y += text_height + 5  # Adjust the line spacing as needed
                                    else:
                                        font_color = (0, 0 ,0)
                                        draw.text((lstlist[i][0], lstlist[i][1]), str(textformat[i]), font=font, fill=font_color)


                        
                            # Save the image with added text
                            output_path = r"C:\Users\dhana\Pictures\session_emotions"+"\\"+"6.png"  # Replace with the desired path and filename
                            image_with_elements.save(output_path)
                        
                        # Display the image with added text
                        image_with_elements.show()





monthtracker().month()
sessionimagescreator().sessionimages()
            
```
